```python
#Librerias
import cv2
import os
import time
import uuid
```

```python
IMAGES_PATH = 'Tensorflow/workspace/images/collectedimages'
```

Creamos la carpeta labels el cual tendra los signos que nos interesan con el numero de imagenes el cual se entrenara

```python
labels = ['hola','si','a']
numero_imagenes = 15
```

creamos la recoleccion

```python
for label in labels:
    !mkdir {'Tensorflow\workspace\images\collectedimages\\'+label}
    cap = cv2.VideoCapture(0)
    print('Recolectando imagenes para {}'.format(label))
    time.sleep(5)
    for imgnum in range(numero_imagenes):
        ret, frame = cap.read()
        imagename = os.path.join(IMAGES_PATH, label, label+'.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imagename, frame)
        cv2.imshow('frame', frame)
        time.sleep(2)

        if cv2.waitKey(1) & 0xff == ord('q'):
            break
    cap.release()    
```
